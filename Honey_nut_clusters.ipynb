#Normalizing:

```python
test_mult = filled_in_df_removed*3
```

```python
test_mult.head()
```

```python
test_mult.iloc[2, 10]
```

```python
def normalize(data_frame):
    for i in range(len(data_frame)):
        data_frame.iloc[i]*1/data_frame.iloc[i, 10]
    return data_frame
```

```python
normalize(filled_in_df_removed)
```
