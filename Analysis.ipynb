```python
import pandas as pd
import numpy as np
```

Loading File

```python
df = pd.read_csv('cereals.csv')
```

```python
cols = list(df.columns)
```

```python
missing = df['carbo']== -1
```

```python
# Replace all -1 with NaNs
for col in cols:
    missing = df[col] == -1
    df[col][missing] = np.nan
```

```python
# replace all the values with mean
df = df.fillna(df.mean())
```

```python
corr_matrix = df.corr()
```

```python
high_corr = corr_matrix > 0.5
```

```python
corr_matrix[high_corr]
```

High Correlation between Protein and Potassium, weight and calories, sugars and calories, potassium and fiber (Minimum correlation greater than 0.5). However the highest of these (> 80) is between Potassium and Fiber at 0.9

```python
#features reduction for clustering. Will look at those with correlation greater than 0.5
cols = ['sugars', 'calories', 'protein', 'fiber', 'potass']
X = df[cols]
Y = df[['name']]
cereal_data = df[df.columns]
```

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
import matplotlib.pyplot as plt
import matplotlib
```

```python
#Scaling the data
scaled_data = preprocessing.scale(X)
```

```python
# clusters
def clusters(no_clusters, data=scaled_data):
    kmeans = KMeans(no_clusters)
    kmeans.fit(data)
    return kmeans.predict(data) # Labels the original data set based on the similarity/closeness between values
```

```python
clusters(3)
```

```python
clusters(4)
```

```python
c5 = clusters(5)
```

```python
clusters(6)
```

```python
clusters(16)
```
