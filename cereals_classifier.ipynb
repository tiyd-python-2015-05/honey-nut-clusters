```python
import pandas as pd
from sklearn.preprocessing import Normalizer
from sklearn import preprocessing
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('cereals.csv', na_values = -1)
```

```python
counter = 0
cval = 0
for col in data.columns:
    for val in data[col]:
        cval += 1
        if val == -1:
            counter += 1
#             print('{}, {}'.format(col,cval))
    print('{} -> {}'.format(col, counter))
    counter = 0 
    cval = 0
```

```python
class Scaled:

    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
#         return X.truediv(X.cups, axis='index')
        cups = X[:,-1]
        cups = cups.reshape(77,1)
        return X / cups
```

```python
estimator = Pipeline([("imputer", preprocessing.Imputer()),
                      ("scaled", Scaled()),
                      ("normalize", Normalizer()),
                      ('kmeans', KMeans(4))])
```

```python
estimator.fit_transform(data)
```

```python
clusters = estimator.predict(data)
```

```python
clus_df = pd.DataFrame(clusters, columns=['cluster'])
```

```python
name_df = pd.DataFrame(names, columns=['name'])
```

```python
df = pd.concat([name_df, clus_df], axis=1)
```

```python
df.sort(columns='cluster')
```

```python
def identifier(clus_num):
    print(df.loc[df['cluster'] == clus_num])
```
